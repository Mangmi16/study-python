```python
from IPython.display import clear_output 

clear_output()

# 숫자를 두개 입력받아 덧셈/뺄셈/곱셈/나눗셈 해주는 사용자 정의함수를 만드셈

def add(n1, n2):
    return n1 + n2

def subtract(n1,n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1,n2):
    return n1/n2

# 연산자를 입력 받아서 (+,-,*,/) 위 함수 중. 연산자에 해당되는 사용정의 함수를 실행하라.
# 사전을 활용해서 키값으로 연산자를 입력 받아  내용인 함수 이름을 출력해준다.
# <- 이 사전을 만들어라. 사전 : "키" : "내용" => 
# 키 = 연산자(+,-,*,/) : 사용자정의함수이름(add,subtract,multiply,divide)
연산자_사전 = {
    "+" : add,
    "-" : subtract,
    "*" : multiply,
    "/" : divide,
}
def 계산기():
    
    no1 = float(input("첫번째 수를 입력하세요."))
    # 연산자를 입력 받아서 (+,-,*,/) 
    for i in 연산자_사전:
        print(i)

    should_continue = True

    while should_continue:
        
        연산자_입력 = input("연산자를 선택하세요: ")
        no2 = float(input("두번째 수를 입력하세요."))

        #연산자_사전[연산자_입력] # 연산자를 받아서 함수를 '지정'해준다.
        연산_함수 = 연산자_사전[연산자_입력] #지정한 함수 이름을 '연산_함수'에 할당을 해준다.
         # 연산함수의 이름으로 함수를 시행한다.
        answer = 연산_함수(no1,no2)
        print(" %s %s %s = %s" %(no1,연산자_입력,no2,answer)) #결과를 출력해 준다


    # 연산이 끝난 후.
    #계산을 계속할 것인지 사용자에게 물어보고 
        if input(f'''현재 값은 {answer} 입니다. 
    현재 값으로 계산을 계속하기 원하시면 'y'를 입력하세요. 
    계산을 처음부터 하고 싶으시면 'n' 을 입력하세요''').lower() == 'y' :
            no1 = answer

        else:
            should_continue = False
            clear_output()



    #y 라고 답 하면 계산을 이어서,
    # n 라고 답하면 출력창을 clear 한 다음 계산을 처음 부터 다시 시작하여라.
    
계산기()
```

```python
#1. 랜덤 단어 생성
word_list = ["apple", "banana", "Cat"]
import random
the_word = random.choice(word_list)



# 라이프와 남은 글자수 표시
life_point = 5
char_left = len(the_word)
print("life point is : ",life_point)


blanks = list("_"* len(the_word))
print("_ "* len(the_word))


while life_point > 0 and char_left > 0 :
    user_input = input("Guess a letter: ")
    correct_indicator = 0

    for i in range(0, len(the_word)):
        #print("current letter: ", the_word[i])

        if user_input.lower() == the_word[i].lower():
            blanks[i] = user_input.lower()
            #print("blank", blanks)
            char_left -= 1
            correct_indicator += 1
        
    if correct_indicator == 0 :
        life_point -= 1
    
    print("current life point is : ",life_point)
    print(blanks)

if life_point == 0:
    print("Dead")
    
else :
    print("Victory!")

        
        
```
