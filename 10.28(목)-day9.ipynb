```python
# 학생정보를 딕셔너리 형태로 입력해주는 함수를 만들어 본다.
def create_student(name, korean, math, english, science):
    return {
        "name" : name,
        "korean" : korean,
        "math" : math,
        "english": english,
        "science" : science,
    }

students = [
    create_student("윤인성", 87, 98, 88, 95),
    create_student("연하진", 92, 98, 96, 98),
    create_student("구지연", 76, 96, 94, 90),
    create_student("나선주", 98, 92, 96, 92),
    create_student("윤아린", 95, 98, 98, 98),
    create_student("윤명월", 64, 88, 92, 92)
]

##*** 학생별로 ***
#1) 총합 구하는 함수
def student_get_sum(student):
    return student["korean"] + student["math"] +\
        student["english"] + student["science"]

#2) 평균 구하는 함수
def student_get_average(student):
    return student_get_sum(student) / 4

#3) 출력해주는 함수
def student_to_string(student):
    return "{}\t{}\t{}".format(
        student["name"],
        student_get_sum(student),
        student_get_average(student))

# 학생을 한 명씩 반복합니다.
print("이름", "총점", "평균", sep="\t")
for student in students:
    # 출력합니다.
    print(student_to_string(student))
```

```python
class Calculator :
    def __init__(self):
        self.AddCount = 0
        self.SubCount = 0
        self.MulCount = 0
        self.DivCount = 0
    def Add (self, num1, num2):
        self.AddCount += 1
        return num1 + num2

    def Sub (self, num1, num2):
        self.SubCount += 1
        return num1 - num2

    def Mul (self, num1, num2):
        self.MulCount += 1
        return num1 * num2

    def Div (self, num1, num2):
        self.DivCount += 1
        return num1 / num2
        
    def ShowCount(self):
        print("덧셈: %s" %self.AddCount)
        print("뺄셈: ", self.SubCount)
        print("곱셈: ", self.MulCount)
        print("나눗셈: ", self.DivCount)
        
        

c = Calculator()

print(c.Mul(100,5))

c.ShowCount()
 
```

# Private variables, getter, setter

* private variable <= 클래스에서만 쓰이는 변수 : 내부의 변수를 외부에서 Access 못하게 할 때

__변수이름
* getter : private variable 값을 가져오는 함수
* setter : private variable 값을 변경해주는 함수


```python
import math
math.pi

class Circle :


    def __init__(self, r):
        self.__radius = r # private variable을 설정해 준다.

        # class 안에 변수 만들기.
        Circle.pi = 3.14

    def area(self):
        return Circle.pi * self.__radius**2

    def circumference(self):
        return Circle.pi * self.__radius * 2

    
    @property
    def radius(self):
        return self.__radius
    
    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise TypeError("양의 숫자만 입력하세요")
        self.__radius = value
        
       
        
circle = Circle(10)
print("반지름:",circle.radius ) # __radius가 프라이빗 변수임에도 불구하고 값을 출력한다
circle.radius = 2
print("setter를 이용해서 변경한 반지름", circle.radius)
```

# 상속

상속은 이미 존재하는 클래스에서 필요한 기능만 가져 오는 것을 의미한다.

* parent(부모)클래스 : 기능을 가져오는 클래스
* child (자식)클래스 : 기능을 이식하는(대상)클래스

다중상속 가능! <= 여러개의 parent class에서 여러개의 기능을 가져 오는것.

override : 상속받아온 메소드

```python
class Country :
    #"parent" class
    name = "국가명"
    population = "인구"
    capital = '수도'

    def show(self):
        print("국가 클래스 메소드.")


class Korea(Country):
    # Child class

    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("국가 이름", self.name)
```

```python
a = Korea('대한민국') # Child 클래스인 Korea를 불러왔다. 
a.show() # a를 child클래스로 할당했음에도 불구하고. parent class 메쏘드인 show() 불러왔다.

a.show_name()

```

```python
class Country :
    #"parent" class
    name = "국가명"
    population = "인구"
    capital = '수도'

    def show(self):
        print("국가 클래스 메소드.")


class Korea(Country):
    # Child class

    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("국가 이름", self.name)
        
```

```python
class Korea(Country):
    # Child class

    def __init__(self, name, population, capital):
        self.name = name
        self.population = population
        self.capital = capital

    def show(self): # parent class에 있는 show()메쏘드와 동일한 이름을 가지고 있다. 
        print(f"""
        이름 : {self.name}
        인구 : {self.population}
        수도 : {self.capital}
        """)
```
